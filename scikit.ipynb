```python
from sklearn import datasets, cluster,metrics
iris=datasets.load_iris()
iris_D=iris.data

result=cluster.KMeans(n_clusters=3).fit(iris_D) #用KMeans把iris_D分群成三類
result.labels_ #列出結果

#用silhouette score算，越接近1，分群結果越好
silhouette_avg=metrics.silhouette_score(iris_D,result.labels_)
print(silhouette_avg)
```

```python
from sklearn import datasets, cluster,metrics
import matplotlib.pyplot as pt
iris=datasets.load_iris()
iris_D=iris.data

ks=range(2,11) #範圍2-10
silhouette_avgs=[]

for n in ks:
    result=cluster.KMeans(n_clusters=n).fit(iris_D) 
    silhouette_avg=metrics.silhouette_score(iris_D,result.labels_)
    silhouette_avgs.append(silhouette_avg)

pt.bar(ks,silhouette_avgs)
pt.show()
```

```python
from sklearn import datasets, cluster, metrics
from sklearn import svm  #svm=支援向量機
from sklearn.cross_validation import train_test_split  #交叉驗證
import matplotlib.pyplot as plt

digits=datasets.load_digits()
#X為資料,Y為分類結果，images為影像檔,三者都要分成train(訓練資料)與test(測試資料)
#75%=0.75為訓練資料，25%=0.25為測試資料，random_state為亂數起始值
X_train, X_test, Y_train, Y_test, images_train, images_test= train_test_split(digits.data, digits.target, digits.images, test_size=0.25, random_state=42)

svc_model=svm.SVC(gamma=0.001, C=100., kernel='linear')
svc_model.fit(X_train,Y_train)
predicted=svc_model.predict(X_test)

#以下為畫圖
images_and_predictions=list(zip(images_test, predicted))
for index, (image, prediction) in enumerate(images_and_predictions[:4]):
    plt.subplot(1,4,index+1)
    plt.axis('off')
    plt.imshow(image, cmap=plt.cm.binary)
    plt.title('Predicted: '+ str(prediction))
plt.show()
```
